Holding some variables constant help normalize a test. _Extrenous variables_ factors are known as _lurking variables_.

Population parameters (such as mu, or $\mu$) are values that describe the entire population. Sample statistics (such as X-bar, or $\overline{x}$) are values that describe our sample; we use statistics to estimate the population parameters. Estimates are our best guesses for the population parameters. So, for example, we would use X-bar to estimate $\mu$.

Some important differences here we need to establish right away. 

* Population $\mu$ is the total population value. 

* Population average $\mu$ is the average for some feature of the total population. 

* Sample average $\overline{x}$ is the sample average. We use sample statistics to approximate population parameters with some sample error. 

The sample statistics can give us an interval where the population parameters lie as long as it's random and unbiased. 

The difference between the population average and the sample average is called the sampling error, denoted as:

Sampling Error = $\mu$ - $\overline{x}$


QUIZ: Below you can find the population score for the memory quiz. Will the average of the sample be equal, greater than, or less than the population sample?

```julia
# Population average versus sample average
using Statistics
population = (100, 82, 54, 98, 93, 74, 32, 61, 48)
mu_population = mean(population)

sample = (100, 98, 93)
mu_sample = mean(sample)

if mu_population > mu_sample
    print("Population average $mu_population is bigger than sample average $mu_sample")
elseif mu_population < mu_sample
    print("Population average $mu_population is smaller than sample average $mu_sample")
else
    print("Population average $mu_population is equal to sample average $mu_sample")
end
```

The above exercise is a good example of how a sample might not estimate very well the population parameter. We can improve the estimation by making the sample larger. 

```julia
# Population average versus sample average with Bigger sample
using Statistics

population = (100, 82, 54, 98, 93, 74, 32, 61, 48)
mu_population = mean(population)

sample = (100, 98, 93, 82, 32, 61)
mu_sample = mean(sample)

if mu_population > mu_sample
    print("Population average $mu_population is bigger than sample average $mu_sample")
elseif mu_population < mu_sample
    print("Population average $mu_population is smaller than sample average $mu_sample")
else
    print("Population average $mu_population is equal to sample average $mu_sample")
end
```

## Visualization of Relationships

## SURVEYS
What are some benefits of using surveys to conduct research?

1. Easy way to get info in a population
2. Relatively inexpensive
3. Conducted remotely
4. Anyone can access & analyze results

There are some downsides. 

1. Untruthful reponses
2. Biased responses
3. Respondents not understanding questions
4. Respondents refuse to answer

Surveys are used to analyze **constructs**. 

## CONTROLLED EXPERIMENTS
Used with control groups and random smaples. The group are sometimes subjected to _blinding_ or the use of placebo. This can be further perfected by the use of the _double blind_ technique where the researcher doesn't know if the patient is part of the control or real group in the test. 

Random assignment of patients works best with large groups. Sometimes researchers use _Within Subject Design_ where the subject is used twice for testing.
